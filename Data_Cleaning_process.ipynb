```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
```

##  DATA CLEANING :

```python
df.isnull().sum()
```

Filling the NAN data with median():

```python
df["Unit_Size_Sqft"]
```

```python
df["Unit_Size_Sqft"] = pd.to_numeric(df["Unit_Size_Sqft"])
```

```python
df.loc[df["Unit_Size_Sqft"] <= 0, "Unit_Size_Sqft"] = None
```

```python
df["Unit_Size_Sqft"] = df["Unit_Size_Sqft"].fillna(df["Unit_Size_Sqft"].median())
```

```python
df.isnull().sum()
```

1.Assign it as string

```python
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].astype(str)
```

```python
df["Ticket_Price_Cr"]
```

2.Replace the str_values:

```python
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].str.replace("₹", "", regex=False)
```

```python
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].str.replace("Cr", "", regex=False)
```

```python
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].str.strip()
```

```python
df["Ticket_Price_Cr"]
```

Now I can change the error value:

```python
df["Ticket_Price_Cr"] = pd.to_numeric(df["Ticket_Price_Cr"], errors='coerce')
```

```python
df["Ticket_Price_Cr"] = df["Ticket_Price_Cr"].fillna(df["Ticket_Price_Cr"].median())
```

```python
df.isnull().sum()
```

```python
df["Amenity_Score"] = df["Amenity_Score"].fillna(df["Amenity_Score"].median())
```

```python
df.isnull().sum()
```

```python
df = df.drop("Buyer_Comments", axis=1)
```

```python
df.isnull().sum()
```

change "Configuration" str_value in same format:

```python
df["Configuration"]
```

```python
df["Configuration"] = df["Configuration"].str.strip().str.upper()
```

```python
df["Configuration"] 
```

Developer Name normalize

Normalize text fields

```python
df["Developer_Name"].head(10)
```

Change the text format as Capt"

```python
df["Developer_Name"] = df["Developer_Name"].str.strip().str.upper()
```

```python
df["Developer_Name"].head(10)
```

```python
df["Micro_Market"] = df["Micro_Market"].str.strip().str.upper()
```

```python
df["Micro_Market"]
```

# Feature Engineering

Price_per_Sqft :

```python
df["Price_per_Sqft"] = (df["Ticket_Price_Cr"] * 10000000) / df["Unit_Size_Sqft"]
```

```python
df["Price_per_Sqft"]
```

```python
df[["Ticket_Price_Cr", "Unit_Size_Sqft", "Price_per_Sqft"]].head()
```

Quarter_Numbers:

Convert Purchase_Quarter to date

```python
df["Purchase_Quarter"] = pd.to_datetime(df["Purchase_Quarter"])
```

Extract Quarter_Number

```python
df["Quarter_Number"] = df["Purchase_Quarter"].dt.quarter
```

Booking_Flag:

Create Booking Flag

```python
df["Booking_Flag"] = df["Transaction_Type"].map({"Primary": 1, "Secondary": 0})
```

```python
df[["Price_per_Sqft", "Quarter_Number", "Booking_Flag"]].head()
```
